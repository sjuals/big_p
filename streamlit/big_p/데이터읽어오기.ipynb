```python
import urllib.request 
import json 
import pandas as pd 

url = 'https://bigdata.gwangju.go.kr/gjAPI/getGpBusiness/getGpBusinesslist.rd?apiSrvCd=0033&pageNo=1&numOfRow=1000'
# 데이터 가져오기
response = urllib.request.urlopen(url) 
json_str = response.read().decode('utf-8')

# JSON으로 변환
json_object = json.loads(json_str)


# 데이터프레임으로 변환
# data = pd.json_normalize(json_object['body']['items'])
data = pd.json_normalize(json_object['response']['body']['items'])

# 확인
data
```

```python
data['price_1'] = data['price_1'].str.replace('원','')
data['price_1'] = data['price_1'].str.replace(',','')
data['price_1'] = data['price_1'].str.strip()
```

```python
data = data.loc[(data['price_1'] !='4000\n(홀 3000)')]
data['price_1'].loc[(data['price_1'] =='')] = 0
```

```python
data['price_1'] = data['price_1'].astype(int)
```
