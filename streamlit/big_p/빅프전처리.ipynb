# 04. K-means clustering (서구 쌍촌동)

```python
df = pd.read_csv('서구 쌍촌동.csv')
```

```python
df.head()
```

```python
df = df[['경도','위도']]
```

```python
sns.lmplot('경도','위도',data=df, fit_reg=False, scatter_kws={'s':20})

plt.title('서구 쌍촌동 카페 좌표')
# plt.xlabel('x')
# plt.ylabel('y')
plt.show()
```

```python
from sklearn.cluster import KMeans

points = df.values
kmeans = KMeans(n_clusters=5).fit(points)
kmeans.cluster_centers_
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for i in range(1, 20):
    kmeans_plus = KMeans(n_clusters=i, init='k-means++')
    kmeans_plus.fit(points)
    inertia.append(kmeans_plus.inertia_)

plt.plot(range(1, 20), inertia, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans.labels_
```

```python
df['cluster_id']= kmeans.labels_
```

```python
sns.lmplot('경도','위도', data=df, fit_reg=False, scatter_kws={'s':20}, hue='cluster_id')
plt.title('서구 쌍촌동 카페 좌표')

plt.show()
```

```python
cc= pd.DataFrame(columns=['경도','위도'], data=kmeans.cluster_centers_)
```

```python
cc
```

```python
sns.lmplot('경도','위도', data= cc, fit_reg=False, scatter_kws={'s':150})
plt.title('쌍촌동 설치 위치')
plt.show()
```

```python
cdf = pd.concat([df,cc],axis=0)
cdf['cluster_id'].fillna('설치위치',inplace=True)
```

```python
# sns.lmplot('x','y', data= cdf, fit_reg=False, scatter_kws={'s':150})
sns.lmplot('경도','위도', data= cdf, fit_reg=False, scatter_kws={'s':20}, hue='cluster_id')
# sns.lmplot('경도','위도', data= cc, fit_reg=False, scatter_kws={'s':30})
plt.title('쌍촌동 카페 및 설치위치')

plt.show()
```
